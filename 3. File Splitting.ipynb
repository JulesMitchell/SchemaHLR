```python
#load dataset (this is dataset created with missing data imputed)
working_dataset = pd.read_excel('workingDataset.xlsx', engine='openpyxl', index_col=[0])

pd.set_option("display.max_rows", 10) #adjust max rows displayed as needed

# define functions
def reverseScoring(df, col): 
    '''Reverse scores on given columns
     df = your data frame,
     col = list of columns you want reversed in list form''' # Create a list for reverse coded items as required
    df[col] = df[col].max() + 1 - df[col]
    return df

# function for adding scale items and calculating mean (Create a list for scale items as required and pass to col)
def scalecalc (df, name, list, divisor):
    '''Calculate sub-scale scores and append column to dataframe
     df = your data frame,
     name = name of column to append (must be a string)
     list = list of columns you want to sum (define lists)
     divisor = For mean, set to length of list otherwise input 1. Note this makes value float
     '''
    df[name] = df[list].sum(axis=1)/divisor
    return df 
```

```python
edited_dataset = working_dataset

# columns added for tallying each mental health diagnosis
edited_dataset['EatD'] = 0 # Eating Disorder'''
edited_dataset['MDD'] = 0 # Major depressive disorder'''
edited_dataset['BiP'] = 0 # Bipolar disorder'''
edited_dataset['SAD'] = 0 # Social anxiety disorder'''
edited_dataset['GAD'] = 0 # Generalised anxiety disorder'''
edited_dataset['PaD'] = 0 # Panic disorder'''
edited_dataset['PTSD'] = 0 # Post traumatic stress disorder'''
edited_dataset['SUD'] = 0 # Substance use disorder'''
edited_dataset['OCD'] = 0 # Obsessive compulsive disorder'''
edited_dataset['BDD'] = 0 # Body dysmorphia disorder'''
edited_dataset['BPD'] = 0 # Borderline Personality disorder'''
edited_dataset['ADHD'] = 0 # Attention deficit hyperactivity disorder'''
edited_dataset['Other'] = 0 # Other'''
edited_dataset['ND'] = 0 # Prefer not to disclose'''
edited_dataset['MHBurden'] = 0 # Mental health burden (i.e. counts)

# string response split (retains string type, hence int() is used in the for loop)
subtypes = edited_dataset['MH_Diag_type'].str.split(',', n=-1, expand=False)

# Checks mental health sub-types, adds value to column if reported, and updates MH burden score
for i in range(len(subtypes)):
    check_participant = subtypes.iloc[i]
    print('participant nr', i)
    try:
        for j in check_participant:
            print(j)
            if int(j) == 1:
                edited_dataset.loc[i, 'EatD'] = 1
                edited_dataset.loc[i, 'MHBurden'] += 1 # this adds 1 to mental health burden column if true. 
            if int(j) == 2:
                edited_dataset.loc[i, 'MDD'] = 1
                edited_dataset.loc[i, 'MHBurden'] += 1 
            if int(j) == 3:
                edited_dataset.loc[i, 'BiP'] = 1
                edited_dataset.loc[i, 'MHBurden'] += 1
            if int(j) == 4:
                edited_dataset.loc[i, 'SAD'] = 1
                edited_dataset.loc[i, 'MHBurden'] += 1
            if int(j) == 5:
                edited_dataset.loc[i, 'GAD'] = 1
                edited_dataset.loc[i, 'MHBurden'] += 1
            if int(j) == 6:
                edited_dataset.loc[i, 'PaD'] = 1
                edited_dataset.loc[i, 'MHBurden'] += 1
            if int(j) == 7:
                edited_dataset.loc[i, 'PTSD'] = 1
                edited_dataset.loc[i, 'MHBurden'] += 1
            if int(j) == 8:
                edited_dataset.loc[i, 'SUD'] = 1
                edited_dataset.loc[i, 'MHBurden'] += 1
            if int(j) == 9:
                edited_dataset.loc[i, 'OCD'] = 1
                edited_dataset.loc[i, 'MHBurden'] += 1
            if int(j) == 10:
                edited_dataset.loc[i, 'BDD'] = 1
                edited_dataset.loc[i, 'MHBurden'] += 1
            if int(j) == 11:
                edited_dataset.loc[i, 'BPD'] = 1
                edited_dataset.loc[i, 'MHBurden'] += 1
            if int(j) == 12:
                edited_dataset.loc[i, 'ADHD'] = 1
                edited_dataset.loc[i, 'MHBurden'] += 1
            if int(j) == 13:
                edited_dataset.loc[i, 'Other'] = 1
                edited_dataset.loc[i, 'MHBurden'] += 1  
            if int(j) == 14:
                edited_dataset.loc[i, 'ND'] = 1
                edited_dataset.loc[i, 'MHBurden'] += 1        
    except:
        print('error')


# columns added for tallying each past eating disorder diagnosis
edited_dataset['pastAN'] = 0 # Past Anorexia Nervosa
edited_dataset['pastBN'] = 0 # Past Bulimia Nervosa
edited_dataset['pastBED'] = 0 # Past Binge Eating Disorder
edited_dataset['partOSFED'] = 0 # Past OSFED
edited_dataset['ND'] = 0 # Prefer not to disclose

# string response split (retains string type, hence int() is used in the for loop)
past_subtypes = edited_dataset['past_ED_subtype'].str.split(',', n=-1, expand=False)

# Checks past ED diagnosis sub-types, adds value to column if reported, and updates MH burden score
for i in range(len(past_subtypes)):
    check_participant_two = past_subtypes.iloc[i]
    print('participant nr', i)
    try:
        for j in check_participant_two:
            print(j)
            if int(j) == 1:
                edited_dataset.loc[i, 'pastAN'] = 1
            if int(j) == 2:
                edited_dataset.loc[i, 'pastBN'] = 1  
            if int(j) == 3:
                edited_dataset.loc[i, 'pastBED'] = 1
            if int(j) == 4:
                edited_dataset.loc[i, 'partOSFED'] = 1
            if int(j) == 5:
                edited_dataset.loc[i, 'ND'] = 1      
    except:
        print('error')
```

```python
''' Issues:The column names still currently need to be manually changed'''

#Prior to splitting the files, the questionnaire sub-scales are calculated.

#First, reverse code items before calculating totals
CFI_R = ['CFI_2', 'CFI_4','CFI_7', 'CFI_9','CFI_11', 'CFI_17'] # create multiple lists or extend
reverseScoring(edited_dataset, CFI_R) # pass the dataset and the list

### Define columns for each scale total and assign to variable ###
'''The sub-scales are first defined, then each scale is assigned 
to the variable 'scales' before calling the scalecalc function'''

#Young positive schema questionnaire
YPSQ_EFSA = ['YPSQ_2','YPSQ_5','YPSQ_49','YPSQ_50','YPSQ_51','YPSQ_52']
YPSQ_S = ['YPSQ_6','YPSQ_7','YPSQ_8','YPSQ_9','YPSQ_10']
YPSQ_EC = ['YPSQ_11','YPSQ_12','YPSQ_13']
YPSQ_O = ['YPSQ_14','YPSQ_15','YPSQ_17','YPSQ_18']
YPSQ_EOS = ['YPSQ_19','YPSQ_20','YPSQ_21','YPSQ_22']
YPSQ_SCRE = ['YPSQ_23','YPSQ_24','YPSQ_25','YPSQ_39','YPSQ_41']
YPSQ_DS = ['YPSQ_26','YPSQ_27','YPSQ_28']
YPSQ_SB = ['YPSQ_30','YPSQ_31','YPSQ_32','YPSQ_33']
YPSQ_HSC = ['YPSQ_34','YPSQ_35','YPSQ_36','YPSQ_37']

#Cognitive flexibility index
CFI_Alt = ['CFI_1','CFI_3','CFI_5','CFI_6','CFI_8','CFI_10','CFI_12','CFI_13','CFI_14',\
    'CFI_16','CFI_18','CFI_19','CFI_20']
CFI_Ctrl = ['CFI_2','CFI_4','CFI_7','CFI_9','CFI_11','CFI_15','CFI_17']
CFI = [CFI_Alt,CFI_Ctrl]

#Emotional regulation index
EmReg_CR = ['EmReg_1','EmReg_3','EmReg_5','EmReg_7','EmReg_8','EmReg_10'] # cognitive reappraisal
EmReg_ES = ['EmReg_2','EmReg_4','EmReg_6','EmReg_9'] # expressive suppression

#Eating Disorder Examination Questionnaire 
EDE_Q_R = ['EDE-Q_1','EDE-Q_2','EDE-Q_3','EDE-Q_4','EDE-Q_5'] #restraint
EDE_Q_W = ['EDE-Q_8','EDE-Q_12','EDE-Q_22-28_1','EDE-Q_22-28_3','EDE-Q_22-28_4'] # weight concern
EDE_Q_S = ['EDE-Q_7','EDE-Q_8','EDE-Q_10','EDE-Q_11','EDE-Q_22-28_2',\
    'EDE-Q_22-28_5','EDE-Q_22-28_6','EDE-Q_22-28_7'] # shape concern
EDE_Q_E = ['EDE-Q_7','EDE-Q_9','EDE-Q_19_1','EDE-Q_20_1','EDE-Q_21_1'] # eating concern
EDE_Q_G = ['EDE-Q_1','EDE-Q_2','EDE-Q_3','EDE-Q_4','EDE-Q_5','EDE-Q_8','EDE-Q_12','EDE-Q_22-28_1','EDE-Q_22-28_3','EDE-Q_22-28_4','EDE-Q_7','EDE-Q_8','EDE-Q_10','EDE-Q_11','EDE-Q_22-28_2',\
    'EDE-Q_22-28_5','EDE-Q_22-28_6','EDE-Q_22-28_7','EDE-Q_7','EDE-Q_9','EDE-Q_19_1','EDE-Q_20_1','EDE-Q_21_1']

# Individual scales assigned to variable. Note CFI scales are separate as they don't require mean
scales = [YPSQ_EFSA,YPSQ_S,YPSQ_EC,YPSQ_O,YPSQ_EOS,YPSQ_SCRE,YPSQ_DS,\
    YPSQ_SB,YPSQ_HSC,EmReg_CR,EmReg_ES, EDE_Q_R,EDE_Q_W,EDE_Q_S,EDE_Q_E,EDE_Q_G]

for list in scales: #This iterates through each list, assigns column name and calculates mean.
    scalecalc(edited_dataset, str(list),list, len(list))

for list in CFI: #This calculates total for CFI subscales and grand total
    scalecalc(edited_dataset, str(list),list, 1)
```
